<h1>Decorator,Property e getter!</h1>

◼ Oque é Decorator em python "@"<br>
◼ Property - Um getter no modo pythonico<br>
◼ getter um método para obter atributo<br>

<h3>Decorator</h3>
 De acordo com a <b>PEP 318</b> O método atual para transformar funções e métodos (por exemplo, declarando-os como classes ou métodos estáticos) é complicado e pode levar a código que é difícil de entender. Idealmente, essas transformações devem ser feitas no mesmo ponto do código onde a própria declaração é feita. Esta PEP introduz uma nova sintaxe para transformações de uma função ou declaração de métodos.<br>

 <h3>Oque podemos entender?</h3>
 No contexto pythonico da programação Orientada a Objeto quando trabalhamos com propriedades da classe sua sintaxe usada para definir as propriedades é muito concisa e legível. Quando a PEP fala de uma nova sintaxe para transformar funções estaticas em algo mais flexivel ela esta se referindo a um metodo que transforma uma caracteristica da classe, ou seja uma função que retorna valores de uma atributo da classe.<br>

 <h3>@property</h3>
 É uma propriedade do objeto, ela é um método que se comporta como um atributo, ao usar @property, você pode "reutilizar" o nome de uma propriedade, alem de evitar quebra de codigo utilizando esse método o codigo se torna mais flexivel a alterações.<br>

 neste conteudo teremos dois exemplos comprovando sua eficiencia!



```python
class Livro:
    def __init__(self, classificacao) -> None:
        self.classificacao = classificacao


clear_code = Livro("Tecnico")
        
print(clear_code.classificacao)
```

No caso a cima temos uma Classe chamada livro onde possuimos uma propriedade chamada classificação onde passamos um atributo a essa propriedade.<br> 
Imagine o seguinte contexto:<br> 
Criamos um programa onde possui varios livros registrados e varias classificações, o programa esta funcionando perfeitamente, mas na equipe em que trabalhamos um dos programadores não esta satisfeito com o nome da tal atributo chamado <b>classificacao</b> alegando que esse nome é muito grande e então resolve muda-lo para <b>tipo</b>.<br>

Oque você acha que acontecerá com o codigo?<br> 
O codigo quebrará correto ?<br>
Como resolvemos esse problema ?

```python
class Livro: 
    def __init__(self,tipo) -> None:
        self.tipo = tipo     

    def get_tipo(self):
        return self.tipo        #<- a propriedade da classe deixa de ser estatico graças ao método
                                # que retorna seu valor independente do nome da propriedade
    

clear_code = Livro("tecnico")

print(clear_code.get_tipo())  #<-chamando o método
```

<h3>Transformando um <b>método</b> em uma <b>Propriedade</b> usando @property</h3> 

```python
class Livro: 
    def __init__(self,tipo) -> None:
        self.tipo = tipo     

    @property
    def get_tipo(self):
        return self.tipo   
       
    
clear_code = Livro("tecnico")

print(clear_code.get_tipo()) #<- preste atenção aqui!
```

no exemplo a cima chamei o método usando "()" para executar nosso método e o programa me retornou esse erro, o objeto n pode ser chamado.
<br>

Quando falei em transformar o método em uma propriedade não estava brincando o @property é responsavel por essa mudança.


```python
class Livro: 
    def __init__(self,tipo) -> None:
        self.tipo = tipo     

    @property
    def get_tipo(self):
        return self.tipo  
       
    
clear_code = Livro("tecnico")

print(clear_code.get_tipo) #<- preste atenção aqui! o método get_tipo agora tem uma caracteristica unica passada pelo @property que é retornar o atributo.
```

Independente do nome da propriedade, o retorno sempre será o mesmo!
<h3>Resumo</h3>
◼Decorator - Ela é usada para dar uma funcionalidade "especial" a certos métodos<br>
◼@property - Modifica o método atribuindo uma caracteristica unica retornando valores de tal propriedade da classe.
